```python
import cv2
import sys

def drawBoundingBoxes(imageData, imageOutputPath, inferenceResults, color):
    """Draw bounding boxes on an image.
    imageData: image data in numpy array format
    imageOutputPath: output image file path
    inferenceResults: inference results array off object (l,t,w,h)
    colorMap: Bounding box color candidates, list of RGB tuples.
    """
    for res in inferenceResults:
        left = int(res['left'])
        top = int(res['top'])
        right = int(res['left']) + int(res['width'])
        bottom = int(res['top']) + int(res['height'])
        label = res['label']
        imgHeight, imgWidth, _ = imageData.shape
        thick = int((imgHeight + imgWidth) // 900)
        print(left, top, right, bottom)
        cv2.rectangle(imageData,(left, top), (right, bottom), color, thick)
        cv2.putText(imageData, label, (left, top - 12), 0, 1e-3 * imgHeight, color, thick//3)
    cv2.imshow('mk',imageData)
    cv2.waitKey(0)
    cv2.imwrite(imageOutputPath, imageData)

def main(imagePath, left, top, width, height, label):
    imgcv = cv2.imread(imagePath)
    color = (0,255,0)
    results = [
        {
            "left": left,
            "top": top,
            "width": width,
            "height": height,
            "label": label
        }
    ]
    drawBoundingBoxes(imgcv, './output.png', results, color)

#if __name__ == "__main__":
#    argv = sys.argv
#    print argv
#    main(argv[1], argv[2], argv[3], argv[4], argv[5], argv[6])
```

```python
seq = iaa.Sequential([
    iaa.OneOf([
        iaa.Flipud(1)
        ,iaa.Affine(scale=(0.5, 1.5))
        ,iaa.PiecewiseAffine(scale=(0.03, 0.06))
        ,iaa.Dropout(p=(0.05,0.15))
        
        
    ])

    ,iaa.SomeOf((2)
                ,[
        iaa.GammaContrast(gamma=(0.5,1.75))
        ,iaa.GaussianBlur(sigma=(0.5,1.5))
        ,iaa.ChannelShuffle()
    
    ])
])

def test_pipline_efect(path):
#     data = pd.read_csv()

    img_base_name = random.choice([os.path.basename(im) for im in glob.glob(f"{path}/*.jpg")])
    img = imageio.imread(os.path.join(path,img_base_name))
    grouped = df1.groupby('image_name')
    groped_name = grouped.get_group(img_base_name)
    bbs_df = groped_name[groped_name.columns[1:5]]
    bbs = BoundingBoxesOnImage.from_xyxy_array(bbs_df,shape=img.shape)
    print(bbs)
    
    img_aug,bbs_aug = seq(image=img,bounding_boxes=bbs)
    print("Original image Class name :",img_base_name)
    ia.imshow(bbs.draw_on_image(img,size=2))
    print("Augmented image Class name :",img_base_name)
    ia.imshow(bbs_aug.draw_on_image(img_aug,size=2))
```

```python
seq_pipline = iaa.Sequential([
    iaa.OneOf([
            iaa.Fliplr(1)
       
             ]),
    
    iaa.SomeOf((1,2),
               [
          iaa.MultiplyAndAddToBrightness(mul=(0.5,1.5),add=(-40.40))
         ,iaa.AddToHueAndSaturation(value=(-30,35))
         ,iaa.RemoveSaturation(mul=(0.3,1))
         ,iaa.GammaContrast(gamma=(0.4,1.3),per_channel=True)
         ,iaa.AdditiveGaussianNoise(scale=(0.02*255,0.15*255))
         ,iaa.Snowflakes()
         ,iaa.GaussianBlur(sigma=(0.5,1))
         ,iaa.RemoveSaturation(mul=(0.25,1))
               ],random_order=True),

                            ])
```

```python
def get_read_images(full_image_files_path):
    for image in glob.glob(f"{full_image_files_path}/*.jpg"):
        yield os.path.basename(image),imageio.imread(image)
def bbs_array_to_df(bbs_aug):
    bbs = bbs_aug.to_xyxy_array()
    bbs_df = pd.DataFrame(bbs,columns=['xmin', 'ymin', 'xmax', 'ymax'])
    return bbs_df
```

```python


import re

how_many_time_auguments_per_images = 4
path_img = "flickr_logos_27_dataset/flickr_logos_27_dataset_images1"

images_files_train = os.path.join(os.getcwd(),path_img)
# label_df = pd.read_csv('widht_height.csv')
label_df = df
grouped = label_df.groupby('image_name')

temp_path = os.path.join(os.getcwd(),"aug")
os.makedirs(temp_path,exist_ok=True)
temp_out_path = os.path.join(temp_path,'train')
os.makedirs(temp_out_path,exist_ok=True)

aug_image_df = pd.DataFrame(columns=label_df.columns)

total_img = label_df['image_name'].unique().size

name_image_dont_aug = []

for name,img in get_read_images(images_files_train):
    for epo in range(how_many_time_auguments_per_images):
    
        group_df = grouped.get_group(name).reset_index()
        group_df = group_df.drop(['index'],axis=1)

        bbs_df = group_df[group_df.columns[1:5]]
        bbs = BoundingBoxesOnImage.from_xyxy_array(bbs_df.values,shape=img.shape)

        image_aug,bbs_aug = seq_pipline(image=img, bounding_boxes=bbs)

        bbs_aug = bbs_aug.remove_out_of_image().clip_out_of_image()

        if re.findall("Image...",str(bbs_aug)) == ['Image([]']:
            name_image_dont_aug.append(name)
            print(name)
            pass
        
        else:
            imageio.imwrite(os.path.join(temp_out_path,f"aug_{epo}_{name}"),image_aug)
            new_df = group_df[['image_name','width','height','class_name']]

            for index,_ in new_df.iterrows():
                new_df.loc[index,'height'] = image_aug.shape[0]
                new_df.loc[index,'width'] = image_aug.shape[1]
                new_df.loc[index,'image_name'] = f"aug_{epo}_{name}"

            bbs_df = bbs_array_to_df(bbs_aug)

            aug_df = pd.concat([new_df,bbs_df],axis=1) #columns added
            aug_image_df = pd.concat([aug_image_df,aug_df])



aug_image_df = aug_image_df.reset_index()
aug_image_df = aug_image_df.drop('index',axis=1)
aug_image_df.to_csv("train_label.csv",index=None)
print(" "*2,"-"*5,f"Total images was {how_many_time_auguments_per_images*total_img+how_many_time_auguments_per_images}","-"*5,end='\n\n')
print(">"*15," Game On ","<"*15)
```
